# Importing all Libraries

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

# Reading File

```python
df = pd.read_csv('StudentsPerformance.csv')
df
```

```python
label = list(df['gender'].value_counts().keys())
label
```

```python
values = list(df['gender'].value_counts().values)
values
```

# Bar Chart

```python
plt.bar(label , values)
```

# Rotation of X-axis Values

```python
plt.bar(label , values)
plt.xticks(rotation = 60)
plt.show()
```

# Changing Color of Chart

```python
plt.hist(df['math score'] , edgecolor = 'black' , color='pink')
plt.show()
```

# Using Seaborn Library (relplot) 

```python
sns.relplot(x = 'math score' , y = 'reading score' , data=df)
```

# Using Hue

```python
sns.relplot(x = 'math score' , y = 'reading score' , data=df , hue='gender')
```

# Using Style

```python
sns.relplot(x = 'math score' , y = 'reading score' , data=df , hue='gender' , style='parental level of education')
```

# Using Col

```python
sns.relplot(x = 'math score' , y = 'reading score' , data=df , hue='gender' , style='parental level of education' , col= 'parental level of education')
```

# Using Kind

```python
sns.relplot(x = 'math score' , y = 'reading score' , data=df , hue='gender' , style='parental level of education' , col= 'parental level of education', kind='line')
```

# Catplot

```python
sns.catplot(x = 'lunch' , y = 'writing score' , data=df)
```

# Using Kind in Catplot

```python
sns.catplot(x = 'lunch' , y = 'writing score' , data=df , kind='bar')
```

# Kind = violin

```python
sns.catplot(x = 'lunch' , y = 'writing score' , data=df , kind='violin')
```

# Using scatterplot in seaborn library

```python
sns.scatterplot(x = 'math score' , y = 'reading score' , data=df)
```

# Using barplot in Seaborn

```python
sns.barplot(x='race/ethnicity' ,y = 'writing score' , data = df ,hue= 'gender')
```

# Bar Chart using Catplot

```python
sns.catplot(x='race/ethnicity' ,y = 'writing score' , data = df ,hue= 'gender' , kind='bar')
```

# Putting Title in Chart

```python
sns.catplot(x='race/ethnicity' ,y = 'writing score' , data = df ,hue= 'gender' , kind='bar')
plt.title('Day 16 Assignment')
```

```python
label = list(df['race/ethnicity'].value_counts().keys())
label
```

```python
values = list(df['race/ethnicity'].value_counts().values)
values
```

# Pie Chart

```python
plt.pie(values , labels = label )
plt.show()
```

# Border Color Changing in Pie Chart

```python
plt.pie(values , labels = label , wedgeprops= {'ec':'k'})
plt.show()
```

# Using Explode in Pie Chart

```python
ls = [0,0,0,0,0.2]
plt.pie(values , labels = label , wedgeprops= {'ec':'k'} , explode=ls)
plt.show()
```

# Showing Percentage in Pie Chart

```python
ls = [0,0,0,0,0.2]
plt.pie(values , labels = label , wedgeprops= {'ec':'k'} , explode=ls , autopct='%1.1f%%')
plt.show()
```

# Showing Title in Pie Chart

```python
ls = [0,0,0,0,0.2]
plt.pie(values , labels = label , wedgeprops= {'ec':'k'} , explode=ls , autopct='%1.2f%%' , shadow=True)
plt.title('Day 16 Assignment')
plt.show()
```
